```python
from ultralytics import YOLO

# Load your trained model
model = YOLO("runs/classify/current_model/weights/best.pt")

# Path to the image you want to test
image_path = "dataset/deck/test/electro spirit/202403171846215_jpg.rf.c68420878f338e6be0b3df253843f39d.jpg"

# Run prediction on the single image
results = model.predict(source=image_path, save=False, verbose=False)

# Print out the prediction
for r in results:
    pred_class = r.names[r.probs.top1]          # predicted class
    conf = r.probs.top1conf.item()              # confidence score
    print(f"Predicted: {pred_class} ({conf:.2f})")
```
